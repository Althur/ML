Tópico: <font color=red>Curva ROC</font>

Interlocutor: <font color=red>Um médico oncologista</font>

Resposta:

algumas notações:
TP: True Positive (positivo verdadeiro)
FP: False Positive (positivo falso)
TN: True Negative (negativo verdadeiro)
FN: False Negative (negativo falso)

Sobre a Receiver Operating Characterístic (ROC)

A primeira coisa que precisamos entender é como chegamos na representação gráfica. 
No lado das abscissas, temos o False Positive Rate(FPR) e, do lado das ordenadas, temos o True Positive Rate(TPR). 
O TPR nada mais é do que o Recall (revocação), ou seja, das ocorrências positivas, quantas foram percebidas?
TPR = TP/(TP+FN)

O FPR visa medir o oposto, ou seja, das ocorências negativas, quantas foram percebidas como positivas?
FPR = FP/(FP+TN)
O FPR também é definido como sendo complementar do True Negative Rate(TNR). Ou seja, FPR = 1-TNR. O TNR é conhecido com o especificidade, que é a fração de negativos verdadeiros que foram identificados como negativos pelo agente classificador. 

Em geral, se faz necessário escolher entre qual métrica dar mais importância. Por exemplo, em um exame preliminar de cancêr de mama é melhor ter um TPR bem alto mesmo que a custas de um FPR alto também. Enquanto num exame para decidir sobre a realização ou não de uma masectomia, seria interessante ter um FPR mais baixo, apesar de que o TPR é o mais importante, o FPR passa a ter muito mais peso do que para o exame preliminar. 

Como sabemos que não vivemos no mundo ideal, sabemos que não é possível ter um resultado perfeito para ambas as métricas. Para FPR o ideal seria ter 0% e para o TPR 100%. Com isso, num gráfico de TPR(y) vs FPR(x), caso tivessemos um classificador completamente aleatório veriamos uma linha reta indo do ponto (0,0) ao (1,1), efetivamente cortando ao meio o quandrado formado pelos valores extremos de cada métrica. E um classificador perfeito seria representado pela reta unindo os pontos (0,0) e (0,1) e (1, 1). Em outras palavras, gostariamos de maximizar a área por de baixo da curva, que no caso máximo seria 1. 

Uma característica curiosa da ROC é que caso tenhamos algum algoritmo/sistema que é pior que um classificador aleatório, ou seja, opera com a área ROC-AUC abaixo de 0.5, uma alterniativa seria justamente inverter os sinais dele e assim obter um resultado melhor.

Tópico: <font color=red>Precision versus recall</font>

Interlocutor: <font color=red>Um professor de GDE</font>

Resposta:

Bom, primeiro precisamos entender por que é de interesse à sociedade este conhecimento.
Uma das principais formas de interagirmos com a sociedade onde vivemos é através do acesso à informação, seja via mídias tradicionais ou não. Somos bombardeados com números, estatísticas que muitas vezes nem se quer faze sentido. 
Estamos presenciando isso diariamente com a pandamia no noticiário. 
Bom, aí que entra entendermos um pouco de estatística simples tanto para possibilitar que leiamos os estudos científicos a respeito da pandemia, como para compreendermos alguns termos usados na mídia. 
Um bem básico, mas interessante ponto é a relação entre o Precision (precisão) e o Recall(revocação). 
A precisão mede, a grosso modo, quanto dos valores descritos como positivos de fato são positivos(TP). 
representado pela equação TP/(TP+FP) . sendo FP os valores erroneamente indentificados como positivos. 
o Recall mede, a grosso modo, o % de valores que foram detectados como positivos dentro do universo dos verdadeiramente positivos. Representado pela equação TP/(TP+FN)


Num mundo ideal(claramente não este), iriamos tentar ter 100% de Precision e 100% de Recall. No entanto, isso não é possível, dado que nenhum modelo é perfeito, sempre haverá um momento em que para se obter um % maior de Precision, por exemplo, haverá uma penalidade significativa no % de Recall. 

Saber a qual dar mais relevância é importante para poder contruir algoritmos bons, e isso depende muito da aplicação.
Vamos a alguns exemplos, digamos que queremos um sistema capaz de nos dizer se alguém está com Covid-19. O que é mais importante, que este modelo (1)acerte o maior número de contaminados ou que (2)mesmo deixando passar alguns contaminados, quando acusase contaminação seria mais certeiro? Digamos que o objetivo deste teste é para que possamos rapidamente saber se uma pessoa está contaminada e poder isolar essa pessoa enquanto as pessoas que estão sem o vírus podem circular livremente. Ora, neste caso, seria interessante darmos ênfase ao (1), certo? 
Bom, claro que na vida real podemos trabalhar com alguns cenários distintos, mas, por efeito de simplicidade, vamos assumir que o melhor neste caso seria a primeira alternativa. Neste caso, deveriamos dar um peso maior para este resultado. Por acaso, este resultado é medido através do Recall. 
No entanto, como podemos ver pelo probleminha descrito, um dos objetivos é deixar as pessoas circularem ambientes públicos com segurança. Por tanto, de nada adianta um algoritmo que diga que todo mundo deve ficar em casa e ponto final, neste sentido, é interessante termos um precision mínimo. 
Com isso, definimos um precision minimo de 50% e um recall de 95%, por exemplo. É possível traçarmos um gráfico com a relação Precision vs Recall, nele seria possível identifcarmos a escala de trade-offs e verificar se atingir estes requisitos mínimos de precision e recall seriam possível, lembrando de dar ênfase para o recall, dada a natureza do problema. 

Outro exemplo legal é a taxa de imunização de uma vacina... será que ela funciona mesmo? Neste caso, o que seria mais importante, Precision ou Recall? 
Claro que do ponto de vista da vacina o mais importante é o Recall, mas, do ponto de vista da ciência por trás, é mais interessante sabermos quando que um paciente foi imunizado de verdade, com o mínimo de falsos positivos, para ter uma noção mais conservadora da efetividade da vacina.

```python
# Divide, de modo estratificado, o conjunto de dados.

split = StratifiedShuffleSplit(
    n_splits=1,
    test_size=0.2,
    random_state=RANDOM_SEED)
for train_index, test_index in split.split(df, df["type"]):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python

zoo_train_set = strat_train_set.drop("type", axis=1)
zoo_train_set = zoo_train_set.drop("animal", axis=1)
zoo_train_labels = strat_train_set["type"].copy()

#zoo_train_num = zoo_train_set.drop("legs", axis=1) # categorical
#zoo_train_cat = zoo_train_set["legs"]

zoo_test_set = strat_test_set.drop("type", axis=1)
zoo_test_set = zoo_test_set.drop("animal", axis=1)
zoo_test_labels = strat_test_set["type"].copy()

#zoo_test_num = zoo_test_set.drop("legs", axis=1) # categorical
#zoo_test_cat = zoo_test_set["legs"]
```

```python
# decision_tree model # classico over-fitting

tree_reg = DecisionTreeRegressor(random_state=RANDOM_SEED)
tree_reg.fit(zoo_train_set, zoo_train_labels)

predicted_labels = tree_reg.predict(zoo_test_set)
print("Predição: {}".format(predicted_labels))
print("Original: {}".format(zoo_test_labels.values))

zoo_predictions = tree_reg.predict(zoo_test_set)
tree_mse = mean_squared_error(zoo_test_labels, zoo_predictions)
tree_rmse = np.sqrt(tree_mse)
print("Regressão linear: RMSE = {:.2f}".format(tree_rmse))
```

```python
lin_reg = LinearRegression()
lin_reg.fit(zoo_train_set, zoo_train_labels)

y_pred = lin_reg.predict(zoo_test_set)
lin_rmse = np.sqrt(mean_squared_error(zoo_test_labels, y_pred))
print("Regressão linear: RMSE = {:.2f}".format(lin_rmse))
```

```python
forest_reg = RandomForestRegressor(n_estimators=10, random_state=RANDOM_SEED)

forest_reg.fit(zoo_train_set, zoo_train_labels)

y_pred = forest_reg.predict(zoo_test_set)
forest_rmse = np.sqrt(mean_squared_error(zoo_test_labels, y_pred))
print("Regressão random forest: RMSE = {:.2f}".format(forest_rmse))
```

```python
# scores lineares
lin_scores = cross_val_score(
    lin_reg,
    zoo_train_set,
    zoo_train_labels,
    scoring="neg_mean_squared_error",
    cv=10,
    n_jobs=-1,
)
lin_rmse_scores = np.sqrt(-lin_scores)

# scores arvore
tree_scores = cross_val_score(
    tree_reg,
    zoo_train_set,
    zoo_train_labels,
    scoring="neg_mean_squared_error",
    cv=10,
    n_jobs=-1,
)
tree_rmse_scores = np.sqrt(-tree_scores)


# scores random forest
forest_scores = cross_val_score(
    forest_reg,
    zoo_train_set,
    zoo_train_labels,
    scoring="neg_mean_squared_error",
    cv=10,
    n_jobs=-1,
)
forest_rmse_scores = np.sqrt(-forest_scores)


def display_scores(scores):
    print("Scores:", scores.round(decimals=2))
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

print('linear')
display_scores(lin_rmse_scores)
print('arvore')
display_scores(tree_rmse_scores)
print('random forest')
display_scores(forest_rmse_scores)
```

```python

param_grid = [
    # try 6 (2×3) combinations of hyperparameters.
    {
        "n_estimators": [5, 10, 15, 30, 40],
        "max_features": [4, 6, 8, 10, 12],
    },
    # then try 4 (1x2×2) combinations with bootstrap set as False.
    {
        "bootstrap": [False],
        "n_estimators": [30, 35, 40, 50],
        "max_features": [3 ,8, 15],
    },
]

forest_reg = RandomForestRegressor(random_state=RANDOM_SEED)

# train across 5 folds, that's a total of (6+4)*5=50 rounds of training.
grid_search = GridSearchCV(
    forest_reg,
    param_grid,
    cv=5,
    scoring="neg_mean_squared_error",
    return_train_score=True,
    n_jobs=-1,
)

t1 = time.perf_counter()
grid_search.fit(zoo_train_set, zoo_train_labels)
t2 = time.perf_counter()

print(f"Tempo gasto: {t2 - t1:.2f} s")
```
